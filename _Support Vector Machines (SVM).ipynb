```python
# Detecting Outliers
def detecting_outliers(df):
    new_df=df.copy()
    dict={}
    new_df=new_df.select_dtypes(include=['int','float'])
    for i in new_df.columns:
        q1=new_df[i].quantile(0.25)
        q3=new_df[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        outliers=new_df[(new_df[i]>lower)|(new_df[i]<upper)][i].tolist()
        dict[i]=outliers
    return dict
    
```

```python
detecting_outliers(df)
```

```python
# Removing Outliers
def removing_outliers(df):
    new_df=df.copy()
    new_df=new_df.select_dtypes(include=['int','float'])
    for i in new_df.columns:
        q1=new_df[i].quantile(0.25)
        q3=new_df[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        new_df=new_df[(new_df[i]>lower)&(new_df[i]<upper)]
    return new_df
        
```

```python
new_df=removing_outliers(df)
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
x_train_scale=scaler.fit_transform(x_train)
```

```python
x_test_scale=scaler.transform(x_test)
```

```python
from sklearn.svm import SVC
```

```python
svc=SVC(kernel='linear')
```

```python
svc.fit(x_train_scale,y_train)
```

```python
y_pred=svc.predict(x_test_scale)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
svc=SVC(kernel='rbf')
```

```python
svc.fit(x_train_scale,y_train)
```

```python
y_pred_1=svc.predict(x_test_scale)
```

```python
accuracy_score(y_pred_1,y_test)
```

```python
# Hyperparameter settings for RBF
C_values = [0.1, 1, 10]
gamma_values = [0.01, 0.1, 1]
```

```python
fig, axes = plt.subplots(len(C_values), len(gamma_values), figsize=(12, 9))
fig.suptitle("RBF Kernel SVM - Effect of C & gamma", fontsize=14)
```

```python
#Use cross-validation to evaluate performance
from sklearn.model_selection import GridSearchCV
```

```python
param_grid = {
    'C': [0.1, 1, 10, 100],
    'gamma': [0.01, 0.1, 1, 10],
    'kernel': ['rbf']
}
```

```python
grid = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    cv=5,               # 5-fold CV
    scoring='accuracy',
    verbose=1,
    n_jobs=-1
)
```

```python
grid.fit(x_train,y_train)
```

```python
print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation Score:", grid.best_score_)
```

```python
best_model = grid.best_estimator_
y_pred = best_model.predict(x_test)
```

```python
print("\nTest Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
