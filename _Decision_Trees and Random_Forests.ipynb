```python
def detecting_outliers(df):
    new_df=df.select_dtypes(include=['int','float'])
    dict={}
    for i in new_df.columns:
        q1=new_df[i].quantile(0.25)
        q3=new_df[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        outliers=new_df[(new_df[i]>upper) | (new_df[i]<lower)][i].tolist()
        dict[i]=outliers
    return dict
    
```

```python
detecting_outliers(df)
```

```python
# Removing Outliers
def removing_outliers(df):
    new_df=df.select_dtypes(include=['int','float'])
    for i in new_df.columns:
        q1=new_df[i].quantile(0.25)
        q3=new_df[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        new_df=new_df[(new_df[i]<=upper) & (new_df>=lower)]
    return new_df
    
    
```

```python
new_df=removing_outliers(df)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
from sklearn.ensemble import RandomForestClassifier
from sklearn import tree
```

```python
x=df.iloc[:,:-1]
y=df.iloc[:,-1]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

```python
D_Tree=DecisionTreeClassifier(max_depth=None)
```

```python
D_Tree.fit(x_train,y_train)
```

```python
plt.figure(figsize=(30,20))
tree.plot_tree(D_Tree)
plt.show()
```

```python
y_pred=D_Tree.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
R_Forest=RandomForestClassifier()
```

```python
R_Forest.fit(x_train,y_train)
```

```python
y_pred_1=R_Forest.predict(x_test)
```

```python
accuracy_score(y_test,y_pred_1)
```
