```python
# Using Iqr Method Finding Ouliers
def detecting_outliers(new_df):
    outliers={}
    for i in new_df.columns:
        q1=new_df[i].quantile(0.25)
        q3=new_df[i].quantile(0.75)
        iqr=q3-q1
        lower=q1-1.5*iqr
        upper=q3+1.5*iqr
        outliers_values=new_df[(new_df[i]>upper)|(new_df[i]<lower)][i].tolist()
        outliers[i]=outliers_values
    return outliers
    
```

```python
# Outliers Are
detecting_outliers(new_df)
```
